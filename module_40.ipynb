### <b>Question No. 1</b>

1. **Histogram**: Used to represent the distribution of a continuous variable. It's helpful for understanding the underlying frequency distribution or pattern of the data.

2. **Bar Plot**: Useful for showing the relationship between a categorical variable and a numerical variable. It can compare quantities across different categories.

3. **Box Plot**: Also known as a box-and-whisker plot, it displays the distribution of a continuous variable across different categories. It helps in identifying outliers and understanding the central tendency and spread of the data.

4. **Violin Plot**: Similar to a box plot, but it also shows the probability density of the data at different values. It's useful for visualizing the distribution of the data and comparing it across different categories.

5. **Scatter Plot**: Used to visualize the relationship between two continuous variables. It helps in understanding the correlation or pattern between the variables.

### <b>Question No. 2

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")

# Plot the line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
plt.title("Signal vs Timepoint for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event")
plt.show()
```

### <b>Question No. 3

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Plot the box plots
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Age distribution across passenger classes')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Fare distribution across passenger classes')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

### <b>Question No. 4

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot the histogram
plt.figure(figsize=(12, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple='stack')
plt.title('Price Distribution of Diamonds by Cut')
plt.xlabel('Price')
plt.ylabel('Count')
plt.legend(title='Cut')
plt.show()
```

### <b>Question No. 6

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix where rows are months, columns are years, and values are passengers
flights_pivot = flights_data.pivot(index="month", columns="year", values="passengers")

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Passenger numbers by month and year")
plt.show()
```
